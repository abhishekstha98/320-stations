```python
import geopandas as gpd
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd

# Load GeoJSON data
nepal_geojson_path = 'C:/Users/USER/Desktop/RESOURCES/ICIMOD PROPOSAL/CODING/LOCATION PLOT/district.geojson'# C:/Users/megat/File_Dell
nepal_gdf = gpd.read_file(nepal_geojson_path)

# Create a NetworkX graph
G = nx.Graph()


df = pd.read_csv("C:/Users/USER/Desktop/RESOURCES/ICIMOD PROPOSAL/CODING/weather_dataset_modified.csv")
stations = df['Location'].unique()
for i, station in enumerate(stations):
    lat = df.loc[df['Location'] == station, 'Latitude'].values[0]
    long =df.loc[df['Location'] == station, 'Longitude'].values[0]
#     print(station, lat, long)
    G.add_node(station, pos=(long,lat))
#     break

# Create a figure and axis
fig, ax = plt.subplots(figsize=(10, 10))

# Plot the GeoJSON map
nepal_gdf.plot(ax=ax, color='white', edgecolor='black')

# Plot the network graph on the same axis
pos = nx.get_node_attributes(G, 'pos')
# print(pos) node_color=range(len(G.nodes)), cmap='viridis',
nx.draw(G, pos, with_labels=False, node_size=50,node_color='red', font_size=5, ax=ax)

# # print(pos) node_color=range(len(G.nodes)), cmap='viridis',
# nx.draw(G, pos, with_labels=False, node_size=100,node_color=range(len(G.nodes)), cmap='plasma', font_size=12, ax=ax)

# plt.title('Network Graph over Nepal Map')
plt.show()
```
